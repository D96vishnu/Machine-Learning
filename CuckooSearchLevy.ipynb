```python
import numpy as np
from Params import params as ps
import fitness as ft
import levyFlight as lf
def main():
    for trial in range(ps.get_trial()):
        np.random.seed(trial)
        
        fitness_list = []
        cs_list = []
        
        #Initial population
        for i in range(ps.get_pop()):
            cs_list.append(lf.Cuckoo_eval())
        
        cs_list = sorted(cs_list, key = lambda X : X.get_fitness())
        BestPosition = cs_list[0].get_position()
        BestFitness = ft.calculate(cs_list[0].get_position())
        for iters in range(ps.get_iterations()):
            
            for n in range(len(cs_list)):
                cs_list[n].get_cuckoo()
                cs_list[n].set_fitness(ft.calculate(cs_list[n].get_position()))
                
                j = np.random.randint(low =0 , high = ps.get_pop())
                if j == n : j = np.random.randint(low =0 , high = ps.get_pop())
                
                if cs_list[n].get_fitness() < cs_list[j].get_fitness() :
                    cs_list[j].set_position(cs_list[n].get_position())
                    cs_list[j].set_fitness(cs_list[n].get_fitness())
            
            cs_list = sorted(cs_list, key = lambda X : X.get_fitness())
            

            for a in range(1,len(cs_list)):
                r = np.random.rand()
                if r < ps.get_pd():
                    cs_list[a].abandon()
                    cs_list[a].set_fitness(ft.calculate(cs_list[a].get_position()))
            
            cs_list = sorted(cs_list, key = lambda X : X.get_fitness())
            
            if cs_list[0].get_fitness() < BestFitness:
                BestFitness = cs_list[0].get_fitness()
                BestPosition = cs_list[0].get_position()
            
            print("trial# :", trial+1)
            print("Iteration# :", iters)
            print("Best Fitness : ", BestFitness)
if __name__ == "__main__":
    main()
```
